## Binomial distribution

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment, for example, when you flip a coin it can by either head or tails.

```python
from scipy.stats import binom

# Define all the probability functions related to this distribution

def dbinom(x, size, prob = 0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Generates random variables from the binomial distribution
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Binomial(10, 0.2) distribution where n = 10 is the number of trials and p = 0.2 is the probability of success in a single trial.
import numpy as np
np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

Write the above functions but for both normal and uniform distributions.

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

def dnorm(x, mean, std):
    """
    Calculates the probability density of the normal distribution
    """
    return norm.pdf(x, loc=mean, scale=std)

def pnorm(x, mean, std):
    """
    Calculates the cumulative probability of the normal distribution
    """
    return norm.cdf(x, loc=mean, scale=std)

def qnorm(p, mean, std):
    """
    Calculates the quantile function of the normal distribution
    """
    return norm.ppf(p, loc=mean, scale=std)

def rnorm(n, mean, std):
    """
    Generates random variables from the normal distribution
    """
    return norm.rvs(loc=mean, scale=std, size=n)

mean = 0
std = 1
n_samples = 2000

print(f"PDF value at x=0 for N(0,1): {dnorm(0, mean, std)}")
print(f"CDF value at x=0 for N(0,1): {pnorm(0, mean, std)}")
print(f"PPF for p=0.5 for N(0,1): {qnorm(0.5, mean, std)}")
random_samples = rnorm(n_samples, mean, std)
print(f"Generate {n_samples} random values from N(0,1): {random_samples}")

fig, axis = plt.subplots(3, 1, figsize=(5, 7))

x = np.linspace(-4, 4, 100)
pdf_values = dnorm(x, mean, std)
axis[0].plot(x, pdf_values, 'r-')

cdf_values = pnorm(x, mean, std)
axis[1].plot(x, cdf_values, 'b-')

axis[2].hist(random_samples, bins=30, density=True, color='g', alpha=0.7)
axis[2].plot(x, pdf_values, 'r-')  # Overlay the theoretical PDF

plt.tight_layout()
plt.show()

from scipy.stats import uniform
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

def dunif(x, start, width):
    """
    Calculates the probability density of the uniform distribution
    """
    return uniform.pdf(x, loc=start, scale=width)

def punif(x, start, width):
    """
    Calculates the cumulative probability of the uniform distribution
    """
    return uniform.cdf(x, loc=start, scale=width)

def qunif(p, start, width):
    """
    Calculates the quantile function of the uniform distribution
    """
    return uniform.ppf(p, loc=start, scale=width)

def runif(n, start, width):
    """
    Generates random variables from the uniform distribution
    """
    return uniform.rvs(loc=start, scale=width, size=n)


start = 0  
width = 1  
n_samples = 2000

print(f"PDF value at x=0.5 for Uniform(0,1): {dunif(0.5, start, width)}")
print(f"CDF value at x=0.5 for Uniform(0,1): {punif(0.5, start, width)}")
print(f"PPF for p=0.5 for Uniform(0,1): {qunif(0.5, start, width)}")
random_samples = runif(n_samples, start, width)
print(f"Generate {n_samples} random values from Uniform(0,1): {random_samples}")

fig, axis = plt.subplots(3, 1, figsize=(5, 7))

x = np.linspace(-0.5, 1.5, 100)
pdf_values = dunif(x, start, width)
axis[0].plot(x, pdf_values, 'r-')

cdf_values = punif(x, start, width)
axis[1].plot(x, cdf_values, 'b-')

axis[2].hist(random_samples, bins=30, density=True, color='g', alpha=0.7)
axis[2].plot(x, pdf_values, 'r-')  # Overlay the theoretical PDF

plt.tight_layout()
plt.show()

```
